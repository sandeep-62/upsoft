```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("E:\\DatUpspot\\data0.txt",header=None)
df2 = pd.read_csv("E:\\DatUpspot\\data9.txt",header=None)
df3 = pd.read_csv("E:\\DatUpspot\\data88.txt",header=None)
df4 = pd.read_csv("E:\\DatUpspot\\data89.txt",header=None)
df5 = pd.read_csv("E:\\DatUpspot\\data109.txt",header=None)
df6 = pd.read_csv("E:\\DatUpspot\\data112.txt",header=None)
```

```python
df1 = df1.T
df2 = df2.T
df3 = df3.T
df4 = df4.T
df5 = df5.T
df6 = df6.T
```

```python
df1 = df1.rename(columns={0: "data1"})
df2 = df2.rename(columns={0: "data2"})
df3 = df3.rename(columns={0: "data3"})
df4 = df4.rename(columns={0: "data4"})
df5 = df5.rename(columns={0: "data5"})
df6 = df6.rename(columns={0: "data6"})
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
print("Data1 unique values :- ", df['data1'].unique())
print("Data2 unique values :- ", df['data2'].unique())
print("Data3 unique values :- ", df['data3'].unique())
print("Data4 unique values :- ", df['data4'].unique())
print("Data5 unique values :- ", df['data5'].unique())
print("Data6 unique values :- ", df['data6'].unique())
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaller = StandardScaler()
x_train = scaller.fit_transform(x_train)
x_test = scaller.transform(x_test)
```

## Logistic Regressor

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
logistic = LogisticRegression()
parameter={'C':[0.5,1,2,3,5,6,7,9,10,30,45],'penalty':['l1','l2','elasticnet']}
logistic_regressor = GridSearchCV(logistic,parameter,scoring="accuracy",cv = 5)
logistic_regressor.fit(x,y)
```

```python
print('best parameter :', logistic_regressor.best_params_)
print("best_score : ",  logistic_regressor.best_score_)
```

```python
logistic_predict = logistic_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("The accuracy score of logistic is - ", accuracy_score(y_test,logistic_predict))
```

## Support vector classification

```python
from sklearn.svm import SVC
svc = SVC(kernel='sigmoid', degree= 6)
svc.fit(x_train,y_train)
svc_predict = svc.predict(x_test)
```

```python
print("Accuracy score of SVC is ", accuracy_score(y_test,svc_predict))
```

## Naiv Bias

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
model_predict = model.predict(x_test)
```

```python
accuracy_score(y_test,model_predict)
```
